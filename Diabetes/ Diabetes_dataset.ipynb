```python
## Data cleaning
```

```python
(df == 0).sum()
```

```python
cols_missing = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
```

```python
# Replace the 0's with NaN

df[cols_missing] = df[cols_missing].replace(0, np.nan)
```

```python
# Replace the NaN with the median values

df.fillna(df.median(),inplace=True)
```

```python
## Loading seaborn and mathploat lib
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Glucose distribution by diabetes status
```

```python
sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title("Glucose Levels vs Diabetes Outcome")
plt.show()
```

```python
## Median glucose is higher in diabetics (Outcome = 1) than non-diabetics.
```

```python
# BMI distribution by diabetes status
```

```python
sns.boxplot(x='Outcome', y='BMI', data=df)
plt.title("BMI vs Diabetes Outcome")
plt.show()
```

```python
## Median BMI is higher for diabetics (Outcome = 1) than non-diabetics.
```

```python
#Correlation heatmap
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Medical Features")
plt.show()
```

```python
## Outcome balance
```

```python
df['Outcome'].value_counts(normalize=True)*100
```

```python
## Histograms / KDE plots to see distribution shapes.
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Glucose'], kde=True, bins=30, color='skyblue')
plt.title('Glucose Level Distribution')
plt.xlabel('Glucose')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Bivariate Analysis - Feature vs Diabetes Outcome
```

```python
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
```

```python

plt.figure(figsize=(15, 8))

for i, col in enumerate(features, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(
        x='Outcome', 
        y=col, 
        data=df, 
        hue='Outcome',      
        palette='Set2', 
        legend=False
    )
    plt.title(f'{col} by Diabetes Outcome')
    plt.xlabel("Diabetes Outcome (0 = No, 1 = Yes)")
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
# Univariate Analysis
```

```python
num_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 
            'SkinThickness', 'Insulin', 'BMI', 
            'DiabetesPedigreeFunction', 'Age']
```

```python
plt.figure(figsize=(15, 20))

for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 2, i)
    sns.histplot(df[col], kde=True, color='skyblue', bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
